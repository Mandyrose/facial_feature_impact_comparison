```python
#data prep:

# change unnamed column to 'pairs'
df.rename(columns = {list(df)[0]: 'pairs'}, inplace = True)
```

```python
#split the tuples to two columns
df_splited = df['pairs'].str.split(',', expand=True)
```

```python
#adding   
df.insert(loc=0, column='img1', value=df_splited[0] )
df.insert(loc=1, column='img2', value=df_splited[1] )
```

```python
# Dropping old Name columns 
df.drop(columns =["pairs"], inplace = True) 
```

```python
#add a colums to detect 'hv' or 'cf'
df.insert(df.shape[1], "cf", True)
df.insert(df.shape[1], "hv", True)
```

```python
#detect what are the cf and hv rows
df['cf']=df['type'].apply(lambda x: 'TRUE' if (x=="diff_pairs" or x=="same_pairs" or x=="high_ps_pairs" or x=="low_ps_pairs") else 'FALSE')
df['hv']=df['type'].apply(lambda x: 'TRUE' if (x=="frontal-ref" or x=="frontal-quarter_left" or x=="frontal-half_left" or x=="diff_pairs" ) else 'FALSE')
```

```python
df
```

```python
#detecthe hv according to the images
img_in_hv = ((df.type=='diff_pairs') & ((df.img1 == "('CM46_2.png'") |(df.img1 == "('CM50_2.png'") | (df.img1 == "('CM51_2.png'") | (df.img1 == "('CM60_2.png'") | (df.img1 == "('CM61_2.png'") | (df.img1 == "('CM62_2.png'") | (df.img1 == "('CM64_2.png'") | (df.img1 == "('CM80_2.png'") | (df.img1 == "('CM81_2.png'") | (df.img1 == "('CM82_2.png'") | (df.img1 == "('CM83_2.png'")))
df.loc[img_in_hv, 'hv']='TRUE'
```

```python
df
```

```python
# average accross index:
result_mean = df.groupby(['img1', 'img2','num_pics','num_ids','type','cf', 'hv']).mean().fc7.reset_index()
result_std = df.groupby(['img1', 'img2','num_pics','num_ids','type','cf']).std().fc7.reset_index()
```

```python
df
# result_mean
```

```python
cf_max_value = result_mean.groupby(['num_ids', 'num_pics', 'cf']).max().fc7.reset_index()
```

```python
cf_max_value = cf_max_value[cf_max_value.cf != 'FALSE']
#only cf rows
df_cf_to_norm = result_mean.copy()    
df_cf_to_norm = df_cf_to_norm[df_cf_to_norm.cf != 'FALSE']
```

```python
#rename index colum:
df_cf_to_norm = df_cf_to_norm.reset_index()
df_cf_to_norm.rename(columns={"index": "orig_index"}, inplace=True)

#rename index colum:
cf_max_value = cf_max_value.reset_index()
cf_max_value.rename(columns={"index": "orig_index"}, inplace=True)
```

```python
# cf_max_value
```

```python
#divide by max:
for index, row in df_cf_to_norm.iterrows():
    print('index: ', index)
    value_to_divide = df_cf_to_norm.iloc[index]['fc7']
    num_ids=df_cf_to_norm.iloc[index]['num_ids']
    num_pics = df_cf_to_norm.iloc[index]['num_pics']
    cf=df_cf_to_norm.iloc[index]['cf']
#     type_ = df_cf_to_norm.iloc[index]['type']
    value_to_divide_by = cf_max_value.query("num_ids==@num_ids & num_pics==@num_pics & cf == @cf")["fc7"]
    normalized_value =  value_to_divide/ value_to_divide_by
    print('value_to_divide: ', value_to_divide)
    print('num_ids: ', num_ids)
    print('num_pics: ', num_pics)
    print('cf: ', cf)
#     print('type_: ', type_)
    print('value_to_divide_by: ', value_to_divide_by)
    print('normalized_value: ', normalized_value)

    df_cf_to_norm.at[index, 'fc7'] =  normalized_value
    
```
