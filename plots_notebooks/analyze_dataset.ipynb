```python
import plotly
import glob
import os
```

```python
path = "/home/administrator/experiments/familiarity/dataset/processed_pretraining_dataset/phase_perc_size/pretraining_{'train': 0.7, 'val': 0.2, 'test': 0.1}"
phases = ['train', 'val', 'test']
```

```python
import plotly.express as px
import pandas as pd
for phase in phases:
    classes = glob.glob(os.path.join(path, phase, '*'))
    cls_imgs = [len(glob.glob(os.path.join(cl,'*'))) for cl in classes]
    df = pd.DataFrame({'imgs': cls_imgs})
    fig = px.histogram(df, x="imgs")
    fig.show()
```
